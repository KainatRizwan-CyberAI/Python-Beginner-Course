```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Creating a Sample DataFrame
np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Value1': np.random.randn(100) * 100,
    'Value2': np.random.randn(100) * 50 + 50,
    'Value3': np.random.randint(0, 100, 100),
    'Date': pd.date_range(start='2023-01-01', periods=100, freq='D')
}
df = pd.DataFrame(data)

# 2. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value1'], label='Value1')
plt.plot(df['Date'], df['Value2'], label='Value2')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Line Plot of Value1 and Value2 over Time')
plt.legend()
plt.show()

# 3. Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value3', data=df, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Value3')
plt.title('Bar Plot of Value3 by Category')
plt.show()

# 4. Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Value1'], bins=20, alpha=0.7, label='Value1')
plt.hist(df['Value2'], bins=20, alpha=0.7, label='Value2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Value1 and Value2')
plt.legend()
plt.show()

# 5. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value1', data=df, palette='Set2')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.title('Box Plot of Value1 by Category')
plt.show()

# 6. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Value1'], df['Value2'], c='blue', alpha=0.5, label='Data Points')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.title('Scatter Plot of Value1 vs Value2')
plt.legend()
plt.show()

# 7. Pair Plot
plt.figure(figsize=(10, 6))
sns.pairplot(df[['Value1', 'Value2', 'Value3']], diag_kind='kde', markers='+')
plt.suptitle('Pair Plot of Value1, Value2, and Value3', y=1.02)
plt.show()

# 8. Heatmap
corr = df[['Value1', 'Value2', 'Value3']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Correlation Matrix')
plt.show()

# 9. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value2', data=df, palette='muted')
plt.xlabel('Category')
plt.ylabel('Value2')
plt.title('Violin Plot of Value2 by Category')
plt.show()

# 10. KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Value1'], shade=True, label='Value1')
sns.kdeplot(df['Value2'], shade=True, label='Value2')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE Plot of Value1 and Value2')
plt.legend()
plt.show()

# 11. Joint Plot
plt.figure(figsize=(10, 6))
sns.jointplot(x='Value1', y='Value2', data=df, kind='scatter', alpha=0.5)
plt.suptitle('Joint Plot of Value1 and Value2', y=1.02)
plt.show()

# 12. Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, palette='pastel')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count Plot of Categories')
plt.show()

# 13. Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Category', y='Value3', data=df, palette='deep')
plt.xlabel('Category')
plt.ylabel('Value3')
plt.title('Swarm Plot of Value3 by Category')
plt.show()

# 14. Facet Grid
g = sns.FacetGrid(df, col='Category', col_wrap=2)
g.map(plt.hist, 'Value1', bins=15, color='m', alpha=0.7)
g.set_titles("{col_name} Category")
plt.suptitle('Facet Grid of Value1 by Category', y=1.02)
plt.show()

# 15. Customizing Plots
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Value1', data=df, label='Value1', color='r')
sns.lineplot(x='Date', y='Value2', data=df, label='Value2', color='g')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Customized Line Plot of Value1 and Value2 over Time', fontsize=16)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

# Explanation
Creating a Sample DataFrame: Generates a DataFrame with random data for visualization.
Line Plot: Plots two time series on the same plot.
Bar Plot: Creates a bar plot for the average values of a category.
Histogram: Plots histograms for two different sets of values.
Box Plot: Displays the distribution of values across different categories.
Scatter Plot: Creates a scatter plot for two variables.
Pair Plot: Uses Seaborn to create a matrix of scatter plots.
Heatmap: Shows the correlation matrix of the DataFrame.
Violin Plot: Displays the distribution of a variable across categories.
KDE Plot: Plots Kernel Density Estimate for two variables.
Joint Plot: Shows a scatter plot with histograms for the margins.
Count Plot: Visualizes the count of each category.
Swarm Plot: Displays data points for different categories.
Facet Grid: Uses Seaborn to create a grid of plots based on categories.
Customizing Plots: Demonstrates various customizations like labels, titles, legend, grid, and rotation of x-ticks.
