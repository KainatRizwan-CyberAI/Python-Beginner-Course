# Explanation:
Static Data:
The data dictionary contains a dataset with features such as name, age, salary, city, and experience.

Initialization:
The EDA class initializes with the DataFrame created from the static data.

Summary Statistics:
The summary_statistics method displays summary statistics for the dataset using describe().

Data Visualizations:
The data_visualizations method includes:
Histograms to show the distribution of numerical features.
Pairplots to show relationships between numerical features.
Boxplots to detect outliers.
Barplots for categorical features.

Correlation Analysis:
The correlation_analysis method displays the correlation matrix and a heatmap to visualize correlations between numerical features.

Insights:
The insights method generates specific insights from the data, such as identifying the person with the highest salary, calculating the average experience, and finding the most common city among the employees.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define static data
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [25, 32, 45, 28, 34],
    'salary': [50000, 60000, 80000, 45000, 70000],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles'],
    'experience': [2, 7, 10, 4, 6]
}

# Convert static data to DataFrame
df = pd.DataFrame(data)

# Define the EDA class
class EDA:
    def __init__(self, dataframe):
        self.df = dataframe
    
    def summary_statistics(self):
        # Display summary statistics
        print("Summary Statistics:")
        print(self.df.describe(include='all'))
        print("\n")
    
    def data_visualizations(self):
        # Distribution of numerical features
        print("Distribution of Numerical Features:")
        self.df.hist(bins=10, figsize=(10, 8))
        plt.tight_layout()
        plt.show()
        
        # Pairplot to show relationships between numerical features
        print("Pairplot:")
        sns.pairplot(self.df.select_dtypes(include=[np.number]))
        plt.show()
        
        # Boxplot to detect outliers
        print("Boxplot:")
        self.df.plot(kind='box', subplots=True, layout=(2,3), figsize=(10, 8))
        plt.tight_layout()
        plt.show()
        
        # Barplot for categorical features
        print("Barplot for Categorical Features:")
        categorical_cols = self.df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            sns.countplot(x=col, data=self.df)
            plt.show()
    
    def correlation_analysis(self):
        # Correlation matrix
        print("Correlation Matrix:")
        corr_matrix = self.df.corr()
        print(corr_matrix)
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
        plt.show()
    
    def insights(self):
        # Generate insights based on the data
        print("Insights:")
        highest_salary = self.df.loc[self.df['salary'].idxmax()]['name']
        print(f"The person with the highest salary is {highest_salary}.")
        
        avg_experience = self.df['experience'].mean()
        print(f"The average experience of the employees is {avg_experience:.2f} years.")
        
        most_common_city = self.df['city'].mode()[0]
        print(f"The most common city among the employees is {most_common_city}.")

# Initialize the EDA class
eda = EDA(df)

# Perform EDA
print("Performing EDA...\n")

# Summary Statistics
eda.summary_statistics()

# Data Visualizations
eda.data_visualizations()

# Correlation Analysis
eda.correlation_analysis()

# Insights
eda.insights()
```
