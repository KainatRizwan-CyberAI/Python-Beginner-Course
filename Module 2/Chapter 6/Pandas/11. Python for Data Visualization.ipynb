Python provides several libraries for data visualization, each with unique features and strengths. Matplotlib, Seaborn, Plotly, and Bokeh are among the most popular libraries used for creating various types of visualizations.

Matplotlib: A flexible and comprehensive library for creating static, animated, and interactive visualizations.
Seaborn: Built on top of Matplotlib, it provides a high-level interface for drawing attractive statistical graphics.
Plotly: An interactive graphing library that supports interactive plots, including 3D plots and animations.
Bokeh: A library for creating interactive and web-ready plots.

# Importing Libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
# Simple line plot with Matplotlib
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title('Matplotlib Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import pandas as pd
# Simple scatter plot with Seaborn
df = pd.DataFrame({'X': x, 'Y': y})
sns.scatterplot(data=df, x='X', y='Y')
plt.title('Seaborn Scatter Plot')
plt.show()
```

```python
# Simple line plot with Plotly
fig = px.line(df, x='X', y='Y', title='Plotly Line Plot')
fig.show()
```

```python
# Simple line plot with Bokeh
p = figure(title='Bokeh Line Plot', x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, legend_label='Line', line_width=2)
show(p)
```
