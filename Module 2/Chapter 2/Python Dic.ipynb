# let's create a simple Python dictionary program that stores information about students, including their names and ages. Users can add new students, update existing information, remove students, and view the list of students.

```python
def print_menu():
    print("1. Add Student")
    print("2. Update Student")
    print("3. Remove Student")
    print("4. View All Students")
    print("5. Exit")

def add_student(students):
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    students[name] = age
    print("Student added successfully.")

def update_student(students):
    name = input("Enter student name to update: ")
    if name in students:
        age = input("Enter new age: ")
        students[name] = age
        print("Student information updated successfully.")
    else:
        print("Student not found.")

def remove_student(students):
    name = input("Enter student name to remove: ")
    if name in students:
        del students[name]
        print("Student removed successfully.")
    else:
        print("Student not found.")

def view_students(students):
    print("Student List:")
    if not students:
        print("No students found.")
    else:
        for name, age in students.items():
            print(f"Name: {name}, Age: {age}")

def main():
    students = {}
    while True:
        print("\nStudent Management System:")
        print_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            add_student(students)
        elif choice == '2':
            update_student(students)
        elif choice == '3':
            remove_student(students)
        elif choice == '4':
            view_students(students)
        elif choice == '5':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
