# let's create a larger example that combines both dictionaries and sets. We'll create a program to manage a simple library system. Users can add books to the library, borrow and return books, view all available books, and view books borrowed by a specific user.

```python
class Library:
    def __init__(self):
        self.books = {}
        self.borrowed_books = {}
    
    def add_book(self, book_name, author):
        if book_name not in self.books:
            self.books[book_name] = author
            print(f"Book '{book_name}' by {author} added to library.")
        else:
            print("Book already exists in library.")
    
    def borrow_book(self, book_name, borrower):
        if book_name in self.books:
            if book_name not in self.borrowed_books:
                self.borrowed_books[book_name] = borrower
                print(f"Book '{book_name}' borrowed by {borrower}.")
            else:
                print("Book already borrowed by another user.")
        else:
            print("Book not found in library.")
    
    def return_book(self, book_name):
        if book_name in self.borrowed_books:
            del self.borrowed_books[book_name]
            print(f"Book '{book_name}' returned successfully.")
        else:
            print("Book not borrowed by any user.")
    
    def view_available_books(self):
        print("Available Books:")
        for book, author in self.books.items():
            print(f"{book} by {author}")
    
    def view_borrowed_books(self, borrower):
        print(f"Books borrowed by {borrower}:")
        borrowed_books = [book for book, borrower_name in self.borrowed_books.items() if borrower_name == borrower]
        if borrowed_books:
            for book in borrowed_books:
                print(book)
        else:
            print("No books borrowed by this user.")

def print_menu():
    print("1. Add Book to Library")
    print("2. Borrow Book")
    print("3. Return Book")
    print("4. View Available Books")
    print("5. View Borrowed Books")
    print("6. Exit")

def main():
    library = Library()
    while True:
        print("\nLibrary Management System:")
        print_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            book_name = input("Enter book name: ")
            author = input("Enter author name: ")
            library.add_book(book_name, author)
        elif choice == '2':
            book_name = input("Enter book name to borrow: ")
            borrower = input("Enter borrower name: ")
            library.borrow_book(book_name, borrower)
        elif choice == '3':
            book_name = input("Enter book name to return: ")
            library.return_book(book_name)
        elif choice == '4':
            library.view_available_books()
        elif choice == '5':
            borrower = input("Enter borrower name to view borrowed books: ")
            library.view_borrowed_books(borrower)
        elif choice == '6':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
