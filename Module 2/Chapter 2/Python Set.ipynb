# Let's create a simple Python program that demonstrates basic operations with sets. This program will allow users to add elements to a set, remove elements, perform set operations like union, intersection, and difference, and view the contents of the set.

```python
def print_menu():
    print("1. Add Element")
    print("2. Remove Element")
    print("3. Perform Union")
    print("4. Perform Intersection")
    print("5. Perform Difference")
    print("6. View Set")
    print("7. Exit")

def add_element(my_set):
    element = input("Enter element to add: ")
    my_set.add(element)
    print("Element added successfully.")

def remove_element(my_set):
    element = input("Enter element to remove: ")
    if element in my_set:
        my_set.remove(element)
        print("Element removed successfully.")
    else:
        print("Element not found in set.")

def perform_union(my_set):
    other_set = set(input("Enter elements for another set (separated by space): ").split())
    union_set = my_set.union(other_set)
    print("Union of sets:", union_set)

def perform_intersection(my_set):
    other_set = set(input("Enter elements for another set (separated by space): ").split())
    intersection_set = my_set.intersection(other_set)
    print("Intersection of sets:", intersection_set)

def perform_difference(my_set):
    other_set = set(input("Enter elements for another set (separated by space): ").split())
    difference_set = my_set.difference(other_set)
    print("Difference of sets:", difference_set)

def view_set(my_set):
    print("Set:", my_set)

def main():
    my_set = set()
    while True:
        print("\nSet Operations:")
        print_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            add_element(my_set)
        elif choice == '2':
            remove_element(my_set)
        elif choice == '3':
            perform_union(my_set)
        elif choice == '4':
            perform_intersection(my_set)
        elif choice == '5':
            perform_difference(my_set)
        elif choice == '6':
            view_set(my_set)
        elif choice == '7':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
