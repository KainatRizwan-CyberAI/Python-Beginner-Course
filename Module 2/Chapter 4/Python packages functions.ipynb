```python
import math
import random
import datetime
import os
import json

def math_operations():
    """Perform some basic mathematical operations using the math package."""
    number = 16
    sqrt = math.sqrt(number)
    factorial = math.factorial(number)
    power = math.pow(number, 2)
    print(f"Square root of {number} is {sqrt}")
    print(f"Factorial of {number} is {factorial}")
    print(f"{number} to the power of 2 is {power}")

def random_operations():
    """Generate random numbers using the random package."""
    rand_int = random.randint(1, 100)
    rand_float = random.random()
    rand_choice = random.choice(['apple', 'banana', 'cherry'])
    print(f"Random integer between 1 and 100: {rand_int}")
    print(f"Random float between 0 and 1: {rand_float}")
    print(f"Random choice from a list: {rand_choice}")

def datetime_operations():
    """Display current date and time using the datetime package."""
    now = datetime.datetime.now()
    print(f"Current date and time: {now}")
    print(f"Current year: {now.year}")
    print(f"Current month: {now.month}")
    print(f"Current day: {now.day}")

def os_operations():
    """Perform some basic OS operations using the os package."""
    current_directory = os.getcwd()
    files = os.listdir('.')
    home_directory = os.path.expanduser('~')
    print(f"Current directory: {current_directory}")
    print(f"Files in current directory: {files}")
    print(f"Home directory: {home_directory}")

def json_operations():
    """Work with JSON data using the json package."""
    data = {
        'name': 'Alice',
        'age': 30,
        'city': 'Wonderland'
    }
    json_string = json.dumps(data)
    parsed_data = json.loads(json_string)
    print(f"JSON string: {json_string}")
    print(f"Parsed JSON data: {parsed_data}")

def main():
    # Perform math operations
    math_operations()
    
    # Perform random operations
    random_operations()
    
    # Perform datetime operations
    datetime_operations()
    
    # Perform OS operations
    os_operations()
    
    # Perform JSON operations
    json_operations()

if __name__ == "__main__":
    main()
```
