```python
# Example 1: while loop
def while_loop_example():
    i = 0
    while i < 5:
        print("Value of i:", i)
        i += 1

# Example 2: for loop with range
def for_loop_range_example():
    for i in range(5):
        print("Value of i:", i)

# Example 3: for loop with iterable
def for_loop_iterable_example():
    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print("Fruit:", fruit)

# Example 4: for loop with enumerate
def for_loop_enumerate_example():
    fruits = ["apple", "banana", "cherry"]
    for index, fruit in enumerate(fruits):
        print(f"Index: {index}, Fruit: {fruit}")

# Example 5: if statement
def if_statement_example(number):
    if number % 2 == 0:
        print(f"{number} is even")
    else:
        print(f"{number} is odd")

# Example 6: if-elif-else statement
def if_elif_else_statement_example(number):
    if number > 0:
        print(f"{number} is positive")
    elif number < 0:
        print(f"{number} is negative")
    else:
        print(f"{number} is zero")

# Example 7: nested if statement
def nested_if_statement_example(number):
    if number >= 0:
        if number == 0:
            print("Number is zero")
        else:
            print("Number is positive")
    else:
        print("Number is negative")

# Example 8: break statement in a loop
def break_statement_example():
    for i in range(5):
        if i == 3:
            break
        print("Value of i:", i)

# Example 9: continue statement in a loop
def continue_statement_example():
    for i in range(5):
        if i == 2:
            continue
        print("Value of i:", i)

# Example 10: pass statement
def pass_statement_example():
    for i in range(5):
        pass  # Placeholder, does nothing

# Call examples
def main():
    print("Example 1: while loop")
    while_loop_example()

    print("\nExample 2: for loop with range")
    for_loop_range_example()

    print("\nExample 3: for loop with iterable")
    for_loop_iterable_example()

    print("\nExample 4: for loop with enumerate")
    for_loop_enumerate_example()

    print("\nExample 5: if statement")
    if_statement_example(7)

    print("\nExample 6: if-elif-else statement")
    if_elif_else_statement_example(-3)

    print("\nExample 7: nested if statement")
    nested_if_statement_example(0)

    print("\nExample 8: break statement in a loop")
    break_statement_example()

    print("\nExample 9: continue statement in a loop")
    continue_statement_example()

    print("\nExample 10: pass statement")
    pass_statement_example()

if __name__ == "__main__":
    main()
```
